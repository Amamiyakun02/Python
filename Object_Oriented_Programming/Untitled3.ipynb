```python
from selenium import webdriver
import time
```

```python
path = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(path)
```

```python
driver.get("https://kodepos.nomor.net/_kodepos.php?_i=desa-kodepos&sby=010000&daerah=Kab.&jobs=Tanah%20Laut")
print(driver.title)
table_element = driver.find_element_by_xpath('//table')
rows = table_element.find_elements_by_tag_name('tr')

for row in rows:
    columns = row.find_elements_by_tag_name('td')
    for column in columns:
        cell_data = column.text
        print(cell_data)
```
