```python
import numpy as np
import pandas as pd

# TV = pd.read_excel('nama_file.xlsx')
def Generator(rasio: int, panjangData: int, angkaTarget: int) -> list[float] :
    total = sum(rasio)
    probabilitas = np.array([x / total for x in rasio])

    num_samples = min(panjangData,len(rasio))

    while True:
        random_number = np.random.choice(range(1,len(rasio) +1 ), size=num_samples,p = probabilitas)
        while len(random_number) < panjangData:
            random_number = np.append(random_number, np.random.choice(range(1,len(rasio) +1 ), size=1,p = probabilitas))
        # print(random_number)
        if sum(random_number) == angkaTarget:
            break

    np.random.shuffle(random_number)
    return random_number
Rasio = [7,7,10,32,44]
TargetValidation = np.array([
    [35,35,40,38,42,40,38,44,40,40,41,43,49,45,35,40,41,38,40,38,37,37,38,41,40,38,40,39,45,42,45,49,36,39,41,45,42,47,40,38,42,38,39,35,42,36,37,34,31,34,40,41],
    [33,37,39,35,42,45,37,45,46,45,38,42,48,46,38,38,45,38,43,38,44,40,37,40,38,43,35,37,43,40,37,49,40,41,36,40,44,40,43,41,40,38,39,37,40,35,41,44,34,42,43,39],
    [35,36,39,37,38,48,35,42,38,39,40,44,49,43,37,39,42,42,40,40,41,38,40,39,40,45,40,40,48,40,41,50,46,40,40,41,40,43,41,40,42,38,44,39,40,45,50,30,34,32,42,43],
    [37,38,38,34,38,40,36,38,40,41,40,45,47,49,33,39,42,40,41,40,41,36,35,38,39,45,46,39,44,44,45,48,44,43,42,45,46,49,45,44,45,40,46,41,43,39,45,32,35,34,44,45]
])

# display(TargetValidation)
ValidationFrame = pd.DataFrame({
    'X1 Validation': TargetValidation[0],
    'X2 Validation': TargetValidation[1],
    'X3 Validation': TargetValidation[2],
    'Y Validation': TargetValidation[3]
})
# display(ValidationFrame)
```

```python
from collections import Counter

def hitung_jumlah_dan_persentase(data):
    counts = Counter(data)
    total_data = len(data)
    percentages = {key: (count / total_data) * 100 for key, count in counts.items()}
    return counts, percentages

def cek_persentase(data, persentase_tentukan):
    _, percentages = hitung_jumlah_dan_persentase(data)
    
    for angka, persentase in percentages.items():
        if persentase != persentase_tentukan[angka]:
            return False
    
    return True

# Contoh penggunaan
data = [5, 4, 4, 5, 5, 5, 5, 5, 4, 5, 4, 5, 5, 4, 1, 5, 5, 2, 5, 4, 5, 1, 4, 4, 2, 5, 4, 5, 5, 5, 5, 5, 1, 2, 5, 5, 5, 4, 3, 4, 4, 5, 4, 4, 4, 4, 1, 4, 4, 5, 5, 4]
persentase_tentukan = {1: 7, 2: 7, 3: 10, 4: 32, 5: 44}

counts, percentages = hitung_jumlah_dan_persentase(data)
print("Jumlah setiap angka:", counts)
print("Persentase setiap angka:", percentages)

hasil_cek = cek_persentase(data, persentase_tentukan)
print("Persentase sesuai dengan yang ditentukan:", hasil_cek)
```
