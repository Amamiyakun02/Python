```python
import numpy as np
import pandas as pd
from typing import List
def Generator(rasio: int, panjangData: int, angkaTarget: int) -> List[float] :
    total = sum(rasio)
    probabilitas = np.array([x / total for x in rasio])

    num_samples = min(panjangData,len(rasio))

    while True:
        random_number = np.random.choice(range(1,len(rasio) +1 ), size=num_samples,p = probabilitas)
        while len(random_number) < panjangData:
            random_number = np.append(random_number, np.random.choice(range(1,len(rasio) +1 ), size=1,p = probabilitas))
        # print(random_number)
        if sum(random_number) == angkaTarget:
            break

    np.random.shuffle(random_number)
    return random_number

X1 = [35,35,40,38,42,40,38,44,40,40,41,43,49,45,35,40,41,38,40,38,37,37,38,41,40,38,40,39,45,42,45,49,36,39,41,45,42,47,40,38,42,38,39,35,42,36,37,34,31,34,40,41]
# dataGenerated = []
# Rasio = [1,1,15,47,36]
# for x in targetArray:
#     Result = Generator(Rasio,10,x)
#     dataGenerated.append(Result)
#     print(f"final: {Result}")
#     print(len(Result))
#     print(sum(Result))

# display(dataGenerated)
```
