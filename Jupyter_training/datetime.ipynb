```python
import datetime
import numpy as np
```

```python
date = datetime.datetime.now()

print(date.year)
print(date.strftime("%A"))
print(date.strftime("%B"))

if date.strftime("%B") == "February":
    print("Hellow")
```

```python
datetim = datetime.datetime(2023, 2,15)
print(datetim)

centu = date.strftime('%C') 

if int(centu) == 20:
    print("This is 20 Century")
else:
    print("is not 20 century")

print(date.strftime('%f'))

num = [np.random.randint(30,799) for i in range(10)]

print(max(num))

triple = [i**3 for i in range(100)]
print(triple)
```

```python
print(num)
print(np.argmax(num))
rand_num = np.array([np.random.randint(1,200) for x in range(60)])
rand_num = rand_num.reshape([3,4,5])
print(rand_num)
print(rand_num.argmax())
for x in rand_num:
    print(x)
```

```python
import matplotlib.pyplot as plt 

bilx = np.array([i for i in range(-50,51)])
bily = pow(bilx,2)
ax = plt.subplot()
plt.subplot(1,2,1)
plt.plot(bilx,bily)

bily = 2 * bilx
plt.subplot(1,2,2)
plt.plot(bilx,bily)
plt.plot(0,100)
ax.axhline(y = 0,color='k')

plt.show()
```
