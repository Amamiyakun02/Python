```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sinusgenerator(amp,fr,tAkh,theta):
    t = np.arange(0,tAkh,0.1)
    y = amp * np.sin(2*fr*t + np.deg2rad(theta))
    return t,y

t,y = sinusgenerator(1,1,4,0)
par1 = r'$ A = $' + str(1) + ' cm, '
par2 = r'$ \omega = ' + str(1)+ r'$ \mathcal{Hz} $' + ', '
par3 = r'$ \theta = ' + str(0) + r'${0} $'
rumus = r'$ \mathcal{Y} = A.sin(2 \omega t + \theta)$'
plt.title('Grafik Sinus\n' + rumus + '\n'+ par1 + par2 + par3)
plt.xlabel('time(second)')
plt.ylabel('frekuensi')

plt.plot(t,y)
plt.show()
```
