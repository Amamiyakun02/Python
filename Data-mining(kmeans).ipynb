```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances
```

```python
xdata = np.array([2,1,3,5,3])
ydata = np.array([3,4,1,2,5])
df = pd.DataFrame({
    'atribut 1' : xdata,
    'atribut 2' : ydata
})    
display(df)
```

```python
# Number of clusters (k)
k = 2

# Fit k-means model
kmeans = KMeans(n_clusters=k, random_state=42)
df['cluster'] = kmeans.fit_predict(df[['atribut 1', 'atribut 2']])

# Display DataFrame with assigned clusters
print("\nData Frame with Clusters:")
display(df)

# Display centroids
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=['atribut 1', 'atribut 2'])
print("\nCentroids:")
display(centroids)

# Visualize the clusters
plt.scatter(df['atribut 1'], df['atribut 2'], c=df['cluster'], cmap='rainbow')
plt.scatter(centroids['atribut 1'], centroids['atribut 2'], marker='X', s=200, c='black')
plt.title('K-Means Clustering')
plt.xlabel('atribut 1')
plt.ylabel('atribut 2')
plt.show()
```

```python
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances

# Data
xdata = [2, 1, 3, 5, 3]
ydata = [3, 4, 1, 2, 5]

# Create DataFrame
df = pd.DataFrame({
    'atribut 1': xdata,
    'atribut 2': ydata
})

# Display DataFrame
print("Data Frame:")
display(df)

# Define centroids
centroid1 = [df['atribut 1'][0], df['atribut 2'][0]]
centroid2 = [df['atribut 1'][1], df['atribut 2'][1]]

# Create centroids DataFrame
centroids = pd.DataFrame({
    'atribut 1': [2,4],
    'atribut 2': [4,1.5]
})

# Choose two rows to calculate Euclidean distance
euclidean_dist = euclidean_distances(df[["atribut 1", "atribut 2"]], centroids[["atribut 1", "atribut 2"]])

print("Euclidean Distance:")
print(euclidean_dist)
```
